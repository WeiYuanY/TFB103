```python
# 視訊分解成圖片
import cv2
import numpy as np

cap = cv2.VideoCapture('./video/test1.mp4')   # play video file
FPS = cap.get(cv2.CAP_PROP_FPS)               # frame per second
F_Count = cap.get(cv2.CAP_PROP_FRAME_COUNT)   # frame count
print(f'FPS : {FPS:.2f} ms, Frame_Count : {F_Count}')


# 獲取寬度
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
# 獲取高度
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(FPS,width,height)


i = 0
while cap.isOpened():
    if i == 282:
        break
    else:
        i = i+1
    (flag,frame) = cap.read()
    fileName = "image" +str(i) +".jpg"
    print(fileName)
    if flag == True:
        cv2.imwrite(fileName,frame,(cv2.IMWRITE_JPEG_QUALITY,100))
print('end')
```

```python
#圖片合成視訊
import numpy as np
import cv2
tracker = cv2.TrackerMIL_create()


img = cv2.imread("1.jpg")

# 寫入物件的建立
videoWrite = cv2.VideoWriter("new.mp4",-1,FPS,(width, height))



for i in range(1,283):
    if i > 1 and i <= 10:
        fileName = "image"+str(i)+".jpg"
        img = cv2.imread(fileName,1)
        videoWrite.write(img)
        cv2.waitKey(10)
        
    if i > 10 and i <= 40:
        fileName = "image"+str(i)+".jpg"
        img = cv2.imread(fileName,1)
        frame = cv2.flip(img, -1)
        cv2.putText(frame, 'FLIP', (200, 60), cv2.FONT_HERSHEY_TRIPLEX,2, (0, 0, 255), 1, cv2.LINE_AA)
        videoWrite.write(frame)
        cv2.waitKey(10)
    
    elif i > 41 and i <= 80:
        fileName = "image"+str(i)+".jpg"
        img = cv2.imread(fileName,0)
        frame = cv2.flip(img, 2)
        cv2.putText(frame, 'GRAY', (200, 60), cv2.FONT_HERSHEY_TRIPLEX,2, (0, 0, 255), 1, cv2.LINE_AA)
        videoWrite.write(frame)
        cv2.waitKey(10)
        

    elif i > 80 and i <= 120:
        fileName = "image"+str(i)+".jpg"
        img = cv2.imread(fileName,0)
        thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 3, 3)
        cv2.putText(thresh, 'Threshold', (200, 60), cv2.FONT_HERSHEY_TRIPLEX,2, (0, 0, 255), 1, cv2.LINE_AA)
        videoWrite.write(thresh)
        cv2.waitKey(10)
        
    elif i > 120 and i <= 160:
        fileName = "image"+str(i)+".jpg"
        img = cv2.imread(fileName, cv2.IMREAD_GRAYSCALE)
        scharrx = cv2.Scharr(img, cv2.CV_64F,1,0)
        scharry = cv2.Scharr(img, cv2.CV_64F,0,1)
        scharrx = cv2.convertScaleAbs(scharrx)   # 轉回 Uint8 
        scharry = cv2.convertScaleAbs(scharry)  
        scharrxy =  cv2.addWeighted(scharrx, 0.5,scharry, 0.5, 0)        
        cv2.putText(scharrxy, 'Scharr', (200, 60), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 1, cv2.LINE_AA)
        videoWrite.write(scharrxy)
        cv2.waitKey(10)
        
    elif i > 160 and i <= 200:
        fileName = "image"+str(i)+".jpg"
        img = cv2.imread(fileName, cv2.IMREAD_GRAYSCALE)
        sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
        sobelx = cv2.convertScaleAbs(sobelx)   # 轉回 Uint8 
        sobely = cv2.convertScaleAbs(sobely)  
        sobelxy =  cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0) 
        cv2.putText(sobelxy, 'Sobel', (200, 60), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 1, cv2.LINE_AA)
        videoWrite.write(sobelxy)
        cv2.waitKey(10)
          
    elif i > 200 and i <= 240:
        fileName = "image"+str(i)+".jpg"
        img = cv2.imread(fileName,0)
        canny = cv2.Canny(img, 30, 150, apertureSize=3, L2gradient=0)
        cv2.putText(canny, 'CANNY', (200, 60), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 1, cv2.LINE_AA)
        videoWrite.write(canny)
        cv2.waitKey(10)
        
    elif i > 240 and i <= 282:
        fileName = "image"+str(i)+".jpg"
        img = cv2.imread(fileName)
        bgr=cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        cv2.putText(bgr, 'RGB2BGR', (200, 60), cv2.FONT_HERSHEY_TRIPLEX,2, (0 ,0 , 255), 1, cv2.LINE_AA)
        videoWrite.write(bgr)
        cv2.waitKey(10)



print('end')
```
